```python
import quandl
import pandas as pd
from config import apikey
```

```python
quandl.ApiConfig.api_key = apikey
```

```python
data = quandl.get_table('WIKI/PRICES', qopts = { 'columns': ['ticker', 'date', 'open','high','low','close','volume'] }, ticker = ['AAPL', 'MSFT','TSLA','DELL','DIS'], date = { 'gte': '2018-01-01', 'lte': '2019-09-01' })
```

```python
data.head()
```

```python
from stockstats import StockDataFrame

```

```python
stock = StockDataFrame.retype(data)
```

```python
# stockData=pd.dataframe()
data['macd']=stock['macd']
```

```python
data['signal']=stock['macds']
data.head()
```

```python
data.iloc[1,8],data.iloc[1,12]
data=data.sort_values(by=['ticker','date'],ascending=False)
```

```python
for i in range(1, len(data)):
    #                          # If the MACD crosses the signal line upward
    if data.iloc[i,8] > data.iloc[i,12] and data.iloc[i-1,8] <= data.iloc[i-1,12]:
        data['suggestion']='BUY'
        data.head()
        #                          # The other way around
    elif data.iloc[i,8] < data.iloc[i,12] and data.iloc[i-1,8] >= data.iloc[i-1,12]:
        data['suggestion']='SELL'
        data.head()

    #                          # Do nothing if not crossed
    else:
        data['suggestion']='HOLD'
        data.head()

```

```python
data.sort_values(by=['suggestion'])
data=data.reset_index()
data
```

```python
from sqlalchemy import create_engine
from sqlalchemy.ext.automap import automap_base
```

```python
engine = create_engine(('postgresql://postgres:{}@localhost:5432/stockmarket').format('postgres'))
# engine = create_engine("sqlite:///Resources/stock.sqlite")
```

```python
Base = automap_base()
Base.prepare(engine, reflect=True)
Base.classes.keys()
conn = engine.connect()
```

```python
Base.metadata.drop_all(engine)   # all tables are deleted

data.to_sql('stockdata', con=engine)
```

```python
df = pd.read_sql("select * from stockdata", conn)
df.set_index('index')
```
